```
import cv2
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# img is a binary image
def plot_ellipse(img):
  M = cv2.moments(img)
  
  c_x = M["m10"] / M["m00"]
  c_y = M["m01"] / M["m00"]

  mu20 = int(M["m20"] / M["m00"]) - c_x**2
  mu02 = int(M["m02"] / M["m00"]) - c_y**2
  mu11 = int(M["m11"] / M["m00"]) - c_x * c_y

  print("Coordinates of Centroid : ({}, {})".format(c_x, c_y))

  theta =  0.5 * math.atan((2 * mu11) / (mu20 - mu02))
  l = math.sqrt(8 * (mu20 + mu02 + math.sqrt(4 * (mu11)**2 + (mu20 - mu02)**2)))
  w = math.sqrt(8 * (mu20 + mu02 - math.sqrt(4 * (mu11)**2 + (mu20 - mu02)**2)))

  ellipse_float = ((c_x, c_y), (l, w), math.degrees(math.pi/2 + theta))
  cv2.ellipse(img, ellipse_float, (255, 255, 255), 2)

  return img
```

```
# Test 
input = cv2.imread('./Images/Test.png', 0)

input_plot = cv2.cvtColor(input, cv2.COLOR_BGR2RGB)
plt.imshow(input_plot)
plt.title('Input')
plt.xticks([]), plt.yticks([])
plt.show()

output = plot_ellipse(input)

output_plot = cv2.cvtColor(output, cv2.COLOR_BGR2RGB)
plt.imshow(output_plot)
plt.title('Output')
plt.xticks([]), plt.yticks([])
plt.show
```
