In this notebook, we will see how to perform basic operations on images and videos using OpenCV

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('machinelearning.jpg',0) #-1 - loads the image as it is
                                           # 0 - loads the image in grayscale
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #display the image correctly
plt.imshow(gray)
plt.xticks([]),plt.yticks([]) #remove the scale displayed on the x and y-axes
plt.show()
```

```python
vid = cv2.VideoCapture(0)
while(True):
    ret,frame = vid.read()                         #We will read the video frame by frame
    col = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)   #The video will be displayed in grayscale
    cv2.imshow('frame',col) 
    if cv2.waitKey(1) & 0xFF == ord('q'):          #Close the window if the user presses 'q' key
        break
vid.release()                                      #Release the capture once the user exits the video
cv2.destroyAllWindows()
```
